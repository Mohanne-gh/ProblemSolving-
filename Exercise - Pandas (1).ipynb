```python
import pandas as pd

data = pd.read_csv("C:/Users/lenovo/Downloads/sample_sales_data.csv")
data
```

```python
greather500 = data[data["TotalAmount"] > 500] 
print (greather500)
```

```python
#List all orders in the "Electronics" category.
categoryElectronics = data[data["Category"] == "Electronics"]
print (categoryElectronics)
```

```python

allorders = data[(data["Category"] == "Clothing") & (data["Status"] == "Completed")]
print (allorders)
```

```python
#Find orders where the Quantity is greater than 5 or the Status is "Pending."

findorders = data[(data["Quantity"] > 5) | (data["Status"] == "Pending")]
print (findorders)
```

```python
TotalAmountBt = data[data["TotalAmount"].between (200,1000)]
print (TotalAmountBt)
```
