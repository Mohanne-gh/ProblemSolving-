```python
import matplotlib.pyplot as plt
```

```python
hours = ["6 AM", "9 AM","12 PM","3 PM","6 PM","7 PM"]
temperature = ["20 C","25 C","30 C","40 C","45 C","50 C"]

plt.plot(hours,temperature)
plt.title("Hours VS Temperature")
plt.xlabel("hours ") 
plt.ylabel("temperature")
plt.show()
```

```python
products = ['Laptops', 'Phones', 'Tablets'] 
sales = [120, 300, 90]
plt.bar (products,sales, color = "red")
plt.xlabel("products")
plt.ylabel("sales")
plt.show()
```

```python
scores = [55,67,73,72,68,90,88,75,60,59,95,77,62,83]

plt.hist(scores,bins=5,color = "blue", edgecolor = "red")
plt.title ("Student Exam Score Distribution")
plt.xlabel("Score Range")
plt.ylabel("Number Of Student")
plt.show()
```

```python
companies = ["Apple", "Samsung", "Huawei", "Others"]
shares = [35, 30, 20, 15]

plt.pie(shares, labels=companies, autopct='%1.1f%%', startangle=140)
plt.title("Smartphone Market Share")
plt.show()
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 87]

plt.scatter(study_hours, exam_scores, color="purple")
plt.title("Study Time vs Exam Score")
plt.xlabel("Hours of Study")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

```python
biils = [45,50,55,47,60,70,68,48,49,55,66,75]

plt.boxplot(biils)
plt.title("Monthly Electricity Bill")
plt.ylabel("Bills($)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data Creation (you can replace this with a real dataset)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Region A Sales': [200, 220, 250, 270, 300, 310, 400, 420, 450, 470, 500, 550],
    'Region B Sales': [150, 180, 190, 200, 230, 250, 300, 320, 340, 360, 400, 430],
    'Region C Sales': [100, 120, 140, 150, 160, 180, 190, 210, 220, 230, 250, 280],
    'Advertising Budget': [50, 60, 55, 65, 70, 75, 85, 90, 95, 100, 110, 120]
}
df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data)
```

```python
print(data)
```

```python
print(df)
```

```python
# Line Plot - Sales Trends Over Months
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 5))
list_of_regions = ["Region A Sales", "Region B Sales", "Region C Sales"]
for region in list_of_regions:
    plt.plot(df["Month"], df[region], label=region)
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
# Bar Chart - Total Sales by Region
TotalSales = df[["Region A Sales","Region B Sales","Region C Sales"]].sum()
TotalSales.plot (kind="bar", figsize = (5,5), color = ["blue","orange","green"])
plt.title("Total Sales By Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show
```

```python
#Scatter
plt.figure(figsize = (5,5))
plt.scatter(df["Advertising Budget"],df["Region A Sales"],color = "purple",edgecolor = "black")
plt.title("Advertising Budget VS Region A Sales")
plt.xlabel("Advertising Budget")
plt.ylabel("Region A Sales")
plt.show()
```
